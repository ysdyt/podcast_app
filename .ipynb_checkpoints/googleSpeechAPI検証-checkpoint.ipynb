```python
import MeCab
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
mecab = MeCab.Tagger ('-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')
#mecab.parse('')#文字列がGCされるのを防ぐ
```

## もろもろの集計値を出す

```python
file = './text/google_speech_api/08_001_NiRed-false_lev-false_samp44k.txt'
word = []
all_word = []

with open(file, 'r', encoding='utf-8') as f:
    text = f.read()
    print("char_len:",len(text.replace('\n', '')))
    
    node = mecab.parseToNode(text.replace('\n', ''))
    target_part = ('名詞')
    while node:
        #print(node.feature.split(",")[0])
        all_word.append(node.surface) #全単語取得
        if node.feature.split(",")[0] in target_part:
            #print(node.surface)
            word.append(node.surface) #名詞のみ取得
        node = node.next

print("重複有りの全単語数:",len(all_word))        
print("重複有りの名詞単語数:",len(word))

print(len(set(all_word)))

all_word_freq_dict = Counter(all_word)
#print(all_word_freq_dict)
print("重複なしの全単語数:", len(all_word_freq_dict))

print(len(set(word)))

word_freq_dict = Counter(word)
print("重複なしの名詞単語数:", len(word_freq_dict))
print(word)
```

## 重複なしで出現頻度の多い順にsortする

```python
word_freq_dict = Counter(word)

word_freq_dict_without_sw = {}
for k,v in word_freq_dict.items():
    if not re.search(r'^[あ-ん]{1,2}$', k): # ひらがな1~2文字の単語も除外
        #print(k,v)
        word_freq_dict_without_sw[k] = v

for k, v in sorted(word_freq_dict_without_sw.items(), key=lambda x: -x[1]):
    print(str(k) + ": " + str(v))
```

## wordcloudを作成する

```python
word_freq_dict = Counter(word)
stop_words = ['音', '化', '井上', 'お願い', 'こと', '今日', 'それ', 'ところ', 'みたい', '話','的','人','何','自分','方','感じ','みんな','私','僕','今','時','中','後','気','番','誰','風','結局', '技術','確か','そうそう','前','白金','鉱業','工業','最近','他','一番','好き','名前','データ','データサイエンティスト','会社','分析','会','回','人たち','ブレインパッド','毎回','逆','内容','俺','じゃなくて','住所','人','一','あなた','データサイエンス', '一緒', '吉田']
min_cnt = 2

word_freq_dict_without_sw = {}
for k,v in word_freq_dict.items():
    #print(k)
    if k not in stop_words and v > min_cnt: # stop wordの除外, 最低出現数の制限
        if not re.search(r'^[あ-ん]{1,2}$', k): # ひらがな1~2文字の単語も除外
            #print(k,v)
            word_freq_dict_without_sw[k] = v
```

```python
wordcloud = WordCloud(
    background_color="white",
    max_words = 500,
    max_font_size=100,
    random_state=2019,
    width=800,
    height=400,
    font_path='/Users/ysdyt/Documents/TakaoPGothic.ttf',
    stopwords=set(stop_words),
    )
plt.figure(figsize=(21, 12))
wordcloud.generate_from_frequencies(word_freq_dict_without_sw)
plt.axis("off")
plt.imshow(wordcloud)
plt.savefig('./fig/no5.png')
```
