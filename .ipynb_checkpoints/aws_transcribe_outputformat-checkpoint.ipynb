## 文字列のカウントなど

```python
import MeCab
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
mecab = MeCab.Tagger ('-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')
#mecab.parse('')#文字列がGCされるのを防ぐ
```

```python
file = './text/aws_transcribe/02_001_NoiRed-true_lev-true_samp44k.txt'
word = []
all_word = []

with open(file, 'r', encoding='utf-8') as f:
    text = f.read()
    print("char_len:",len(text))

    node = mecab.parseToNode(text)
    target_part = ('名詞')
    while node:
        #print(node.feature.split(",")[0])
        all_word.append(node.surface) #全単語取得
        if node.feature.split(",")[0] in target_part:
            #print(node.surface)
            word.append(node.surface) #名詞のみ取得
        node = node.next

print("重複有りの全単語数:",len(all_word))
print("重複有りの名詞単語数:",len(word))

print(len(set(all_word)))

all_word_freq_dict = Counter(all_word)
print("重複なしの全単語数:", len(all_word_freq_dict))

print(len(set(word)))

word_freq_dict = Counter(word)
print("重複なしの名詞単語数:", len(word_freq_dict))

print(word)
```

```python
word_freq_dict = Counter(word)

word_freq_dict_without_sw = {}
for k,v in word_freq_dict.items():
    if not re.search(r'^[あ-ん]{1,2}$', k): # ひらがな1~2文字の単語も除外
        #print(k,v)
        word_freq_dict_without_sw[k] = v

for k, v in sorted(word_freq_dict_without_sw.items(), key=lambda x: -x[1]):
    print(str(k) + ": " + str(v))
```
